# Plot with mesh deformation {#ref_tutorials_plot_deformed_mesh}

This tutorial shows different commands for plotting a deformed mesh
without data.

A mesh is represented in DPF by a
`MeshedRegion<ansys.dpf.core.meshed_region.MeshedRegion>`{.interpreted-text
role="class"}. You can store multiple `MeshedRegion` objects in a DPF
collection called
`MeshesContainer<ansys.dpf.core.meshes_container.MeshesContainer>`{.interpreted-text
role="class"}.

You can obtain a `MeshedRegion` by creating your own from scratch or by
getting it from a result file. For more information, see the
`ref_tutorials_create_a_mesh_from_scratch`{.interpreted-text role="ref"}
and `ref_tutorials_get_mesh_from_result_file`{.interpreted-text
role="ref"} tutorials.

PyDPF-Core has a variety of plotting methods for generating 3D plots
with Python. These methods use VTK and leverage the
[PyVista](https://github.com/pyvista/pyvista) library.


# Load data to plot


```python
import ansys.dpf.core as dpf
from ansys.dpf.core import examples, operators as ops

# Download and get the path to an example result file
result_file_path_1 = examples.download_piston_rod()

# Create a model from the result file
model_1 = dpf.Model(data_sources=result_file_path_1)
```

# Get the deformation field

To deform the mesh, a nodal 3D vector field specifying the translation
of each node in the mesh is needed.

The following DPF objects can represent such a field and are accepted as
inputs for the `deform_by` parameter of all plot methods:

- A `Field<ansys.dpf.core.field.Field>`{.interpreted-text role="class"}
- A
  `FieldsContainer<ansys.dpf.core.fields_container.FieldsContainer>`{.interpreted-text
  role="class"}
- A result object
- An `Operator<ansys.dpf.core.dpf_operator.Operator>`{.interpreted-text
  role="class"}

Here, we use the
`displacement<ansys.dpf.core.operators.result.displacement.displacement>`{.interpreted-text
role="py:class"} operator, which outputs a nodal 3D vector field of
distances. One can get the operator from the
`Model<ansys.dpf.core.model.Model>`{.interpreted-text role="class"} with
the data source already connected. For more information about extracting
results from a result file, see the
`ref_tutorials_import_data`{.interpreted-text role="ref"} tutorials
section.


```python
disp_op = model_1.results.displacement()

# Define the scale factor to apply to the deformation
scl_fct = 2.0
```

# Plot a deformed model {#ref_plot_deformed_mesh_with_model}

Plot the overall mesh loaded by the model with
`Model.plot()<ansys.dpf.core.model.Model.plot>`{.interpreted-text
role="py:meth"}. To add deformation, pass the displacement operator to
the `deform_by` argument.

:::: note
::: title
Note
:::

The `DpfPlotter<ansys.dpf.core.plotter.DpfPlotter>`{.interpreted-text
role="class"} displays the mesh with edges, lighting and axis widget
enabled by default. You can pass additional PyVista arguments to all
plotting methods to change the default behavior (see options for
[pyvista.plot()](https://docs.pyvista.org/api/plotting/_autosummary/pyvista.plot.html#pyvista.plot)).
::::


```python
model_1.plot(deform_by=disp_op, scale_factor=scl_fct)
```

# Plot a single mesh {#ref_plot_deformed_mesh_with_meshed_region}

## Get the mesh


```python
meshed_region_1 = model_1.metadata.meshed_region
```

# Plot the deformed mesh using `MeshedRegion.plot()`

Use the
`MeshedRegion.plot()<ansys.dpf.core.meshed_region.MeshedRegion.plot>`{.interpreted-text
role="py:meth"} method and pass the displacement operator to the
`deform_by` argument.


```python
meshed_region_1.plot(deform_by=disp_op, scale_factor=scl_fct)
```

# Plot the deformed mesh using `DpfPlotter`

1.  Create an instance of
    `DpfPlotter<ansys.dpf.core.plotter.DpfPlotter>`{.interpreted-text
    role="class"}.
2.  Add the mesh to the scene using
    `add_mesh()<ansys.dpf.core.plotter.DpfPlotter.add_mesh>`{.interpreted-text
    role="py:meth"}, passing the displacement operator to the
    `deform_by` argument.
3.  Render and show the figure using
    `show_figure()<ansys.dpf.core.plotter.DpfPlotter.show_figure>`{.interpreted-text
    role="py:meth"}.


```python
plotter_1 = dpf.plotter.DpfPlotter()
plotter_1.add_mesh(meshed_region=meshed_region_1, deform_by=disp_op, scale_factor=scl_fct)
plotter_1.show_figure()
```

You can also plot data contours on a deformed mesh. For more
information, see `ref_tutorials_plot_contour`{.interpreted-text
role="ref"}.


# Plot several meshes {#ref_plot_deformed_mesh_with_meshes_container}

## Build a collection of meshes

Use the
`split_mesh<ansys.dpf.core.operators.mesh.split_mesh.split_mesh>`{.interpreted-text
role="class"} operator to split the mesh based on the material of each
element. This operator returns a
`MeshesContainer<ansys.dpf.core.meshes_container.MeshesContainer>`{.interpreted-text
role="class"} with meshes labeled according to the split criterion. For
more information, see the `ref_tutorials_split_mesh`{.interpreted-text
role="ref"} and
`ref_tutorials_extract_mesh_in_split_parts`{.interpreted-text
role="ref"} tutorials.


```python
meshes = ops.mesh.split_mesh(mesh=meshed_region_1, property="mat").eval()
print(meshes)
```

# Plot the deformed meshes

Use
`MeshesContainer.plot()<ansys.dpf.core.meshes_container.MeshesContainer.plot>`{.interpreted-text
role="py:meth"} and pass the displacement operator to `deform_by`. This
plots all `MeshedRegion` objects in the `MeshesContainer` and colors
them based on the split criterion.


```python
meshes.plot(deform_by=disp_op, scale_factor=scl_fct)
```

You can also plot data on a collection of deformed meshes. For more
information, see `ref_tutorials_plot_contour`{.interpreted-text
role="ref"}.
