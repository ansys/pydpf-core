# Extract a mesh in split parts {#ref_tutorials_extract_mesh_in_split_parts}

`Fluent`{.interpreted-text role="bdg-fluent"} `CFX`{.interpreted-text
role="bdg-cfx"}

This tutorial shows how to extract meshes split on a given space or time
from a result file.

To accomplish this goal, use the
`meshes_provider<ansys.dpf.core.operators.mesh.meshes_provider.meshes_provider>`{.interpreted-text
role="class"} operator. The split meshes are given in a
`MeshesContainer<ansys.dpf.core.meshes_container.MeshesContainer>`{.interpreted-text
role="class"} and can be spatially or temporally varying.


# Import the necessary modules


```python
from ansys.dpf import core as dpf
from ansys.dpf.core import examples, operators as ops
```

# Define the `DataSources`

Create
`DataSources<ansys.dpf.core.data_sources.DataSources>`{.interpreted-text
role="class"} objects so the `meshes_provider` operator can access the
mesh. For this tutorial, use result files available in the
`examples<ansys.dpf.core.examples>`{.interpreted-text role="mod"}
module. For more information on importing result files, see the
`ref_tutorials_import_data`{.interpreted-text role="ref"} tutorials
section.


```python
# Fluent
result_file_path_3 = examples.download_fluent_axial_comp()["flprj"]
ds_3 = dpf.DataSources(result_path=result_file_path_3)

# CFX
result_file_path_4 = examples.download_cfx_mixing_elbow()
ds_4 = dpf.DataSources(result_path=result_file_path_4)
```

# Extract all mesh parts --- Fluent

Instantiate and evaluate the `meshes_provider` operator.


```python
meshes_31 = ops.mesh.meshes_provider(data_sources=ds_3).eval()
print(meshes_31)
```

Extract all mesh parts --- CFX


```python
meshes_41 = ops.mesh.meshes_provider(data_sources=ds_4).eval()
print(meshes_41)
```

# Scope the mesh regions to extract --- Fluent

A region corresponds to a zone for Fluid and CFX results. Specify the
mesh regions you want to get by passing zone ids to the `region_scoping`
argument.


```python
meshes_32 = ops.mesh.meshes_provider(data_sources=ds_3, region_scoping=[3, 12]).eval()
print(meshes_32)
```

Scope the mesh regions to extract --- CFX


```python
meshes_42 = ops.mesh.meshes_provider(data_sources=ds_4, region_scoping=[5, 8]).eval()
print(meshes_42)
```
